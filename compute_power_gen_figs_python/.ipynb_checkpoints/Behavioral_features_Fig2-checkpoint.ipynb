# Large inter-subject variability in extracted behavioral features

Takes in 10 behavioral and environmental metadata measures and plots their distributions for every subject. This will generate Fig. 2 from the paper.

```python
%matplotlib inline
import glob, natsort, mne, sys, pdb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import gridspec
import seaborn as sns
from mne.time_frequency import read_tfrs
import config
from feat_plot_utils import update_behavioral_features,feat_plot_unit_conversions,feat_plot,plot_feature_scattermatrix
```

### Set parameters

```python
tfr_lp = '.../naturalistic_reaches/tfr/'
n_subjs = 12
feature_labels = ['day','tod','reach_duration','reach_r','reach_a','onset_velocity',
                  'audio_ratio','I_over_C_ratio','other_reach_overlap','bimanual']
```

### Load behavioral/environmental features into 1 dataframe

```python
for i in range(n_subjs):
    fnames_all = natsort.natsorted(glob.glob(tfr_lp+'subj_'+str(i+1).zfill(2)+'*.h5'))

    for s,fname in enumerate(fnames_all):
        #Load in metadata
        power = read_tfrs(fname)[0]
        df_sbj_meta_orig = update_behavioral_features(power.metadata)
        df_sbj_meta = df_sbj_meta_orig.loc[:,feature_labels]
        del power,df_sbj_meta_orig
        df_sbj_meta['patient_id'] = 'subj_'+str(i+1).zfill(2)
        
        #Append to 1 dataframe
        if (s==0) & (i==0):
            df_metadata = df_sbj_meta.copy()
        else:
            df_metadata = pd.concat((df_metadata,df_sbj_meta),ignore_index=True)
            
#Transform data into more interpretable units
df_metadata_plot = feat_plot_unit_conversions(df_metadata)
```

### Plot feature distributions

```python
feat_plot(df_metadata_plot,feature_labels,n_subjs)
```

```python
plot_feature_scattermatrix(df_metadata_plot)
```
